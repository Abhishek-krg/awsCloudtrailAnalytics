```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("event_history_90days.csv")
df.head()
```

```python
df["Event source"]=df["Event source"].apply(lambda x: x.split('.')[0])
```

```python
def plotDfCountPlot(dfCol,xsize,ysize):
    plt.figure(figsize=(xsize,ysize))
    d=dfCol.value_counts()
    ax=sns.barplot(x=d.keys(),y=d.values)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
plotDfCountPlot(df["Event name"],54,26)
```
